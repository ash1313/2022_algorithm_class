(1) 이 알고리즘의 반환값에 대한 순환 관계식을 만들고 풀어라. 이것은 무슨 알고리즘인가?

 $T(n) = T(n-1) + 2n -1, F(1) = 1

 T(n) = T(n-2) + 2(n-1)-1 +2n-1 = T(n-2) + 4(n-1)
 T(n) = T(n-3) + 4(n-1)-4 +2n-1 = T(n-3) + 6n-9
 이 과정을 i번 반복하면
 T(n) = T(n-1-i) + 2(i+1)n -(i+1)^2
 i = n-2을 대입
 T(1) + n^2 -1
 T(1) = 1이므로
 T(n) = n^2
 따라서 n^2을 구하는 알고리즘이다.

(2) 기본 연산으로 곱셈을 사용했을 때의 복잡도를 순환 관계식으로 만들고 풀어라. 시간 복잡도를 설명하라.

 T(n) = T(n-1)+1, T(1) = 0 ,O(n)
 T(n) = T(n-1)+1
      = (T(n-2)+1)+1 = T(n-2)+2
      = (T(n-3)+1)+2 = T(n-3)+3
  이 과정을 i번 반복하면
 T(n) = T(n-i)+i
 i = n-1 을 대입
 T(n) = T(1)+(n-1)
 T(n) = n-1
 
 시간 복잡도 :  O(n)
 

(3) 기본 연산으로 덧셈과 뺄셈을 사용했을 때의 복잡도를 순환관계식으로 만들고 풀어라. 시간 복잡도를 설명하라.

 T(n) = T(n-1)+2, T(1) = 0 ,O(n)
 
 T(n) = T(n-1)+2, T(1) = 0 ,O(n)
 T(n) = T(n-1)+2
      = (T(n-2)+2)+2 = T(n-2)+4
      = (T(n-3)+2)+4 = T(n-3)+6
  이 과정을 i번 반복하면
 T(n) = T(n-i)+2i

[문제 3] 교재 연습문제 133쪽의 문제 5
 무작위로 리스트에 0과 1을 각각 n개씩 넣고 이 리스트에 대해 전체 정렬하는 과정이 출력되도록 프로그램하시오.


```python
#랜덤모듈을 임포트한다.
import random
#a리스트를 생성한다.
a = [0,0,1,1,1,0,0,1,1,0,1,0]
# a = []
# a리스트에 무작위로 리스트에 0,1를 넣는다.
# for value in range(0,random.randint(5,9)):
#   a.append(random.randint(0,1))
#초기값을 출력한다,
print("초기값",a)
# a리스트에 개수를 b에 저장한다.
b = len(a)
change = 1
num = 0
count = 0
while change:
  change = 0
  # 자리 수 만큼 반복한다.
  for i in range(1,b) :
    #연산 횟수를 확인하기 위한 연산
    count = count+1 
    # 1,0이 연속적으로 나오면 
    if a[i-1] == 1 and a[i] ==0 :
      # 0,1 로 바꾼다.
      a[i-1],a[i] = 0,1
      #change로 다시 반복한다.
      change = 1
      num=num+1   
      #리스트가 정렬된 과정을 보여준다.
      print("step",num,a)
#연산 횟수를 출력한다.      
print("연산 횟수 :",count)
#결과를 출력한다.
print("결과 :", a)
```
